```python
### посчитаем показатель ROI - прибыльность вкладов/инвсетиций (выручка:затрачено)
```

```python
df
```

```python
### возьем исходную таблицу, сгруппированную по байерам (их 5), а затем просуммируем нужные столбцы
```

```python
buyer_df_ad = df.groupby('Buyer')['Spend', 'Revenue_ad1d', 'Revenue_ad7d', 'Revenue_ad30d'].sum()
buyer_df_ad
```

```python
### посчитаем ROI за 3 промежутка по выручке с рекламы
```

```python
buyer_df_ad['ROI_1d'] = (buyer_df_ad['Revenue_ad1d'] - buyer_df_ad['Spend']) / buyer_df_ad['Spend']
buyer_df_ad['ROI_7d'] = (buyer_df_ad['Revenue_ad7d'] - buyer_df_ad['Spend']) / buyer_df_ad['Spend']
buyer_df_ad['ROI_30d'] = (buyer_df_ad['Revenue_ad30d'] - buyer_df_ad['Spend']) / buyer_df_ad['Spend']
buyer_df_ad.reset_index()
```

```python
### как видно, везде ROI отрицательный, кроме ROI_30d у 3-его байера, но и он меньше 1. (100%)
```

```python
### далее посчитаем также ROI но уже по покупкам пользователей
```

```python
buyer_df_app = df.groupby('Buyer')['Spend', 'Revenue_in_app1d', 'Revenue_in_app7d', 'Revenue_in_app30d'].sum()
buyer_df_app
```

```python
buyer_df_app['ROI_1d'] = (buyer_df_app['Revenue_in_app1d'] - buyer_df_app['Spend']) / buyer_df_app['Spend']
buyer_df_app['ROI_7d'] = (buyer_df_app['Revenue_in_app7d'] - buyer_df_app['Spend']) / buyer_df_app['Spend']
buyer_df_app['ROI_30d'] = (buyer_df_app['Revenue_in_app30d'] - buyer_df_app['Spend']) / buyer_df_app['Spend']
buyer_df_app.reset_index()
```

```python
### Из таблицы следует, что ROI везде отрицательное значение, а следовательно - инвестировать не имеет смысла
```

```python
### Затем посчитаем общий ROI по 3 промежуткам
```

```python
buyer_fdf = df.groupby('Buyer')['Spend', 'Revenue_ad1d', 'Revenue_ad7d', 'Revenue_ad30d', 'Revenue_in_app1d', 'Revenue_in_app7d', 'Revenue_in_app30d'].sum()
buyer_fdf
```

```python
buyer_fdf['Revenue1d'] = (buyer_df_app['Revenue_in_app1d'] + buyer_df_ad['Revenue_ad1d'])
buyer_fdf['Revenue7d'] = (buyer_df_app['Revenue_in_app7d'] + buyer_df_ad['Revenue_ad7d'])
buyer_fdf['Revenue30d'] = (buyer_df_app['Revenue_in_app30d'] + buyer_df_ad['Revenue_ad30d'])
buyer_fdf
```

```python
buyer_fdf['ROI_1d'] = (buyer_fdf['Revenue1d'] - buyer_fdf['Spend']) / buyer_fdf['Spend']
buyer_fdf['ROI_7d'] = (buyer_fdf['Revenue7d'] - buyer_fdf['Spend']) / buyer_fdf['Spend']
buyer_fdf['ROI_30d'] = (buyer_fdf['Revenue30d'] - buyer_fdf['Spend']) / buyer_fdf['Spend']
buyer_fdf
```

```python
### Из таблицы видно, что ROI везде отрицательный, кроме байера №3, но и там до 1(100%) далеко.
```

```python
### Сгруппируем исходную таблицу по айди кампании, а затем просуммируем значения каждого столбца (расходы и установки)
```
